# Всякое про функции Уолша

Вспомогательные вещи и имя переменной для символьных вычислений

```python
import sympy
from sympy.plotting import plot
sympy.init_printing()

t = sympy.Symbol('t', real = True)
```

Функция Радемахера $r_n = \text{sign} \left(  \sin(2^{n+1}\pi t) \right)$

```python
def radem(n):
    return sympy.sign(sympy.sin(2**(n+1)*sympy.pi*t))
```

Функции Уолша задаём в явном виде (без вычисления степеней) на основе функцтй Радемахера

```python
def walsh(n):
    if n == 0:
        return 1
    if n == 1:
        return radem(0)
    if n == 2:
        return radem(1)
    if n == 3:
        return radem(0)*radem(1)
    if n == 4:
        return radem(2)
    if n == 5:
        return radem(0)*radem(2)
    if n == 6:
        return radem(1)*radem(1)
    if n == 7:
        return radem(0)*radem(1)*radem(2)
    raise ValueError
```

```python
plot(walsh(7),(t,0,1))
plot(walsh(7).subs(t,t/2),(t,0,2))
```
